```python
import numpy as np
import matplotlib.pyplot as plt
from typing import Tuple, List
from sklearn.cluster import KMeans
```

```python
def compress_img(img: np.ndarray, k : int )->np.ndarray:
    X = img.reshape((-1,3))
    kmeans = KMeans(n_clusters=k,n_init='auto')
    y = kmeans.fit_predict(X)
    img_red = kmeans.cluster_centers_[y].reshape(img.shape).astype(np.uint8)
    return img_red
```

```python
img = plt.imread('data/kente.jpg')
imgs = [compress_img(img, k) for k in range(1,33,2)]
```

```python
def greatest_factor(n :int)-> Tuple[int,int]:
    sqrt_n = int(np.sqrt(n))
    max_factor =  1
    for f in range(1, sqrt_n+1):
        if n % f == 0:
            max_factor = f
    other_factor = int(n/max_factor)
    return (max_factor,other_factor)
```

```python
def display_imgs(imgs:List[np.ndarray], starting_k:int=1, k_step =1):
    k = len(imgs)
    nrows, ncols = greatest_factor(k)
    fig, axis = plt.subplots(nrows= nrows, ncols= ncols)
    fig.set_size_inches(10,8)
    for i in range(nrows):
        for j in range(ncols):
            n = (i*ncols)+j 
            p = axis[i,j]
            p.axis('off')
            p.set_title(f'k = {(n*k_step)+starting_k}')
            p.imshow(imgs[n])
    plt.show()
```

```python
display_imgs(imgs, starting_k =1, k_step=2)
```
